```python
import numpy as np
import pandas as pd
```

```python
df = pd.DataFrame(np.random.rand(7,7),
                'a,b,c,d,e,f,g'.split(','), 
                  'S,U,V,W,X,Y,Z'.split(','))
df
```

```python
purvini = df[df>0.1]
purvini
```

```python
purvini.isna() #grąžina lentelę, kur true reikšmės yra NaN (iesko NaN)
```

```python
purvini.isnull() #grąžina lentelę, kur true reikšmės yra NaN
```

```python
purvini.isnull().sum() # Rodo kiek kuriame stulpelyje NaN reikšmių
```

```python
purvini['Y'].isnull() # Galima tikrinti atskiruose stulpeliuose
```

```python
purvini.dropna() # Jis išmeta visas eilutes, kuriose yra bent viena NaN reikšmė:
```

```python
# jeigu norime išmesti stulpelius, turinčius bent vieną NaN, turime nurodyti ašį:

purvini.dropna(axis=1)
```

```python
# Galime išmesti tik tas eilutes, ar stulpelius, kurie turi mažiau, 
# negu (pvz) 5 NE NaN reikšmes:
purvini.dropna(thresh=5)
```

```python
purvini.dropna(thresh=3, axis=1) # axis nuroda kad stulpelius
```

```python
purvini.fillna('kiauras') # vietoj "NaN" rasom "kiauras"
```

```python
# pakeiciam neegzistojami duomeny stulpeliu vidurkiu, tik stulpelis S
print(purvini['S']) # a eilute NaN
purvini['S'].fillna(value=purvini['S'].mean()) # a elute NaN - bet priskiriam stulpelio vidurki
```

```python
print(purvini)

purvini.fillna(value=purvini.mean()) # visur kur NaN, priskiriam jo stulelio vidurkis
```

```python
purvini.fillna(purvini.mean()) # tas pats, kaip priestai, visur kur NaN, priskiriam jo stulelio vidurkis
```

## DF Jungimas

```python
mtop = miestai[:6] #[0:6]
mtop
```

```python
mbot = miestai[6:]
mbot
```

```python
pd.concat([mtop,mbot])
```

```python
pd.concat([mtop,mbot], axis=1) # galime prijungti lentelę iš dešinės:
```

```python
mbot
mright = mbot.reset_index()
mright.drop(columns='index', inplace=True)
mright
```

```python
pd.concat([mtop, mright], axis=1) # pridedam id desines
```

```python
# Pridėkime papildomų eilučių:

# lenkija = {'Šalis': ['Lenkija', 'Lenkija', 'Lenkija'], 
#            'Miestas':['Varšuva', 'Vroclavas', 'Gdanskas'], 
#           'Gyv': [1688, 638, 461]}
# pl = pd.DataFrame(lenkija)
# data_right1 = pd.concat([mright, pl])
```

```python
# Pridėkime papildomų eilučių:

lenkija = {
    'Šalis': ",".join(["Lenkija" for x in range(3)]).split(","),
     'Miestas': 'Varšuva,Vroclavas,Gdanskas'.split(","), 
        'Gyv': [1688, 638, 461]
        }
popolnia = pd.DataFrame(lenkija)
popolnia
```

```python
mpl = pd.concat([mright, popolnia]).reset_index().drop(columns='index') # pridejom lentele prie esamos
mpl
```

```python
pd.concat([mtop, mpl], axis=1)
```

```python
key = 'A,B,C,D,E,F'.split(',')
key
```

```python
key = 'A,B,C,D,E,F'.split(',') # i abi lenteles prideojom sutampancia stulpele 'key', kad ateitije galima sujungti lenteles pagal'key'
mtop['key'] = key
mpl['key'] = key
```

```python
# matome, kad lentelės suklijuotos, tačiau stulpelis 'key' nesikartoja (stulpeliai pervadinti, kad nesidubliuotų pavadinimai). 
# Stulpelį 'key' teko nurodyti parametre on.
pd.merge(mtop, mpl, on = 'key')
```

```python
key2 = 'B,C,D,E,F,G'.split(',')
mtop['key2'] = key2
mpl['key2'] = key2
```

```python
pd.merge(mtop, mpl, on = ['key','key2'])
```

```python
keyX = 'W C D E F W'.split() # keyX = 'W, C, D, E, F, W'.split(',)
mpl['key2'] = keyX
mtop, mpl
```

```python
mtop.join(mpl, lsuffix='L', rsuffix='R') # Yra ir metodas .join(). Jis daro tą patį, ką ir merge, tik indekso pagrindu. Pvz.:
```

```python
mtop.join(mpl, lsuffix='_xxx', rsuffix='_yyy') 
```

```python
mtop.join(mpl, lsuffix='_KAIRE-((1))', rsuffix='_DESINE-((2))').join(mpl)
```
