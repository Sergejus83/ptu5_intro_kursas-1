## .apply() metodas

```python
top50['Length.'].sum()
```

```python
# su .apply() galime maudoti ir savo funkcijas. Parašykime funkciją, kuri iš duotų sekundžių skaičiaus išskirs minutes ir sekundes formatu mm:ss
def minutes_seconds(seconds):
    minutes = seconds // 60
    seconds = seconds % 60
    return f'{minutes}:{seconds:02d}' # 02d - 0- uzpildo tuscia vieta, 2 -minimum 2 skaiciai "d"-sketmuo
```

```python
print(minutes_seconds(121))
print(minutes_seconds(3))
```

```python
top50['Length.'].apply(minutes_seconds)
```

```python
top50['Length.Pretty'] = top50['Length.'].apply(minutes_seconds)
top50.head()
```

```python
# Galime naudoti lambdas: energy duomenys dalinam is 10
top=top50
top['Energy'].apply(lambda x: round(x/10)).head()
```

```python
top['Length.'].apply(lambda seconds: f'{seconds // 60}:{(seconds % 60):02d}')
```
